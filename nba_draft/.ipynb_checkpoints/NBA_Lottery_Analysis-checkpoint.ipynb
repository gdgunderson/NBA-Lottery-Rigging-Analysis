# Is the NBA Lottery Fair? 

This year, NBA fans witnessed what may go down as the most shocking trade in league history. Luka Dončić—a 25-year-old franchise player and five-time All-Star—was traded by the Dallas Mavericks to the Los Angeles Lakers, less than a year after leading his team to the NBA Finals. 

Mavericks fans were irate. Not only did they lose their star player in the prime of his career, but the team didn't seem interested in cashing in on Doncic's full value. General Manager Nico Harrison, who orchestrated the trade and later defended it to the press, became the focal point of fan outrage. Throughout the remainder of the season, chants of “Fire Nico” echoed through the Mavericks’ arena.

![Davis & Doncic](https://hoopshype.com/wp-content/uploads/sites/92/2025/02/Luka_-AD_Trade.png)

As the playoffs began, coverage of the trade had started to fade—until it was brought back into sharp focus: the Dallas Mavericks won the first overall draft pick with just a 1.8% chance. Cooper Flagg, widely projected as a generational talent and the consensus top pick, would be headed to Dallas—if they kept the pick.

NBA fans were quick to speculate: was the trade part of a larger plan? Did the Mavericks trade Dončić for Anthony Davis because they were somehow promised Cooper Flagg in return? It seemed implausible that a team could make one of the most lopsided trades in recent memory… and then immediately win the draft lottery against overwhelming odds.

While this could be pure coincidence, the move made sense from the league's broader perspective. The NBA has faced declining viewership over the last decade, and placing a generational star like Doncic in a major market like Los Angeles could reinvigorate fan interest. It raises a provocative question:

Could the NBA draft be rigged?

**Null Hypothesis (H₀):**
The observed NBA Draft Lottery results are consistent with a fair lottery process. Any deviations from expected outcomes are due to random chance.

**Alternative Hypothesis (H₁):**
The observed NBA Draft Lottery results are not consistent with a fair lottery process—some teams may have received more favorable outcomes than expected by chance alone.

````python
import requests
import os
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import glob
from collections import defaultdict
````

## Web Scraping NBA Draft Lottery Results
This code iterates through each year from 1985 to 2025, sending HTTP requests to RealGM’s draft lottery results pages. For each year:
- It retrieves the HTML content and parses it with BeautifulSoup.
- It locates the draft lottery results table using flexible class matching.
- It extracts the table headers and rows into a DataFrame.
- A "Year" column is added to tag the data.

The data is saved as a CSV file named nba_draft_lottery_<year>.csv.

````python
# Ensure the data folder exists
os.makedirs("../data", exist_ok=True)

# Scrape data and save to data/ folder at project root
for year in range(1985, 2026):
    url = f"https://basketball.realgm.com/nba/draft/lottery_results/{year}"
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        tables = soup.find_all('table')
        print(f"[{year}] Found {len(tables)} tables on the page.")
        
        table = soup.find('table', class_='tablesaw') or \
                soup.find('table', class_='data') or \
                soup.find('table')
        
        if table:
            headers = [th.text.strip() for th in table.find_all('th')]
            rows = [
                [td.text.strip() for td in tr.find_all('td')]
                for tr in table.find_all('tr')[1:]
            ]
            
            df = pd.DataFrame(rows, columns=headers)
            df['Year'] = year
            
            df.to_csv(f"../data/nba_draft_lottery_{year}.csv", index=False)
            print(f"Saved: ../data/nba_draft_lottery_{year}.csv ({len(rows)} rows)")
        else:
            print(f"No table found for {year}")
    else:
        print(f"Failed to fetch {year} (Status: {response.status_code})")
    
    time.sleep(2)
````

## NBA First Pick Deviation Analysis (1990–2025)
This plot visualizes how often teams won the 1st overall pick in the NBA Draft relative to their pre-lottery odds.

Deviation Score: 
- Measures how much a team’s actual 1st pick frequency deviates from expected (based on lottery odds). Higher values indicate "luckier" outcomes.

Key Annotations:
- Team labels highlight notable drafts (e.g., Cavaliers in 2003/2014, Pelicans in 2019).
- Red dotted lines mark years with lottery reforms (1995, 2019), which adjusted odds to discourage tanking.

# Simulating the Lottery from 1990 to 2025
In 1990 a weighted lottery system was introduced. Teams with a worse regular season performance the prior year had a higher chance of drawing earlier draft picks. We will only include data from years since the weighted system was introduced.

The code below collects official draft results from 1990 to 2025, removes unnecessary columns, and formats the data appropriately to perform a Monte Carlo simulation analysis.

````python
# get data since 1990 drop unnecessary cols
real_data = df_all[df_all['Year'] >= 1990].reset_index(drop=True)
real_data = real_data.drop(['Record', 'Chances', 'Pre-Lottery Position', 'Pick Change', 'Draft Team'], axis=1)
real_data.head()
````

````python
# convert 'Odds' column to float
real_data['Odds'] = real_data['Odds'].str.replace('%', '').astype(float) / 100
real_data.head()
````

## Constructing Yearly Team Odds
The code below creates a dictionary called yearly_odds that maps each year to a nested dictionary of teams and their corresponding lottery odds for that year.
For each year, a dictionary is created where: 
- Keys are team name
- Values are their lottery odds in that specific year

```The result is a structure like:
{
    1990: {'Team A': 0.25, 'Team B': 0.20, ...},
    1991: {'Team C': 0.14, 'Team D': 0.12, ...},
    ...
}
```

````python
yearly_odds = {}

# group dataframe by year and create nested dictionaries
for year, group in real_data.groupby('Year'):
    yearly_odds[year] = dict(zip(group['Team'], group['Odds']))

print(yearly_odds)
````

# What is a Monte Carlo Simulation?
Monte Carlo simulations are a statistical technique used to model the probability of different outcomes when there is uncertainty in a system. By running a process many times, we can observe the distribution of possible results and compare them to real-world outcomes.

In this experiement, we use Monte Carlo simulations to model how the NBA Draft Lottery would play out over time if it were completely fair (based purely on each team’s odds). Comparing the actual results to this distribution helps us evaluate whether the real outcomes appear consistent with chance or potentially suspicious.

## Simulation Function
This function simulates NBA draft lotteries across multiple years (yearly_odds) by running num_simulations (default: 100,000) trials. For each simulation, it generates a draft order for every year by randomly selecting teams based on their lottery odds (adjusted to avoid zeros) and records the results in a DataFrame. The output is a list of DataFrames, each representing a full simulation of draft picks for all years.

````python
np.random.seed(42)

def run_multi_year_simulation(yearly_odds, num_simulations=100000):
    """
    Simulates the lottery across all years simultaneously
    
    Returns:
    - List of DataFrames (each representing one full simulation based through every year in 'yearly_odds')
    """
    all_simulations = []

    # loop over the number of simulations
    for _ in range(num_simulations):
        sim_results = []

        # for each year fin yearly odds
        for year, odds_dict in yearly_odds.items():
            teams = list(odds_dict.keys())
            probs = np.array(list(odds_dict.values()), dtype=np.float64)
        
            # normalize and fix zero probs if needed
            if np.count_nonzero(probs) < len(teams):
                probs[probs == 0] = 1e-8  # assign a small value to zero-prob teams
            probs = probs / np.sum(probs)

            # make choice
            draft_order = np.random.choice(
                teams,
                size=len(teams),
                replace=False,
                p=probs
            )
            
            # record results
            sim_results.append(pd.DataFrame({
                'Year': year,
                'Pick': range(1, len(teams)+1),
                'Team': draft_order,
                'Simulated': True
            }))
        
        all_simulations.append(pd.concat(sim_results))
    
    return all_simulations
````

## Comparing Real Outcomes vs. Simulated Outcomes

The analyze_results function evaluates how unusual the real NBA draft results are by comparing them to thousands of simulated lottery outcomes. It uses Monte Carlo analysis to model expected outcomes under fair lottery odds, which vary by year.

The function first counts how often each team received each pick in the real and simulated data. It then calculates a fairness score for each team-pick pair, based on how much the real result deviates from the simulation average, normalized by expected variance.

To assess overall fairness, it computes a chi-squared-like deviation between real and expected results. It then compares this to the distribution of deviations from the simulations, generating an empirical p-value that reflects how extreme the real results are.

Monte Carlo simulation is used instead of a standard chi-squared test because the lottery system’s complexity makes traditional assumptions unreliable.

The difference between a Monte Carlo simulation and a classical chi-squared goodness-of-fit test lies in the distribution used to evaluate the actual results. The classical chi-squared test compares the observed outcomes to a theoretical chi-squared distribution, which is only valid if certain assumptions (such as independence and sufficient expected counts) are met. In contrast, a Monte Carlo simulation creates an empirical distribution based on repeated simulations under fair conditions, allowing for more flexibility when those assumptions are violated.

````python
def analyze_results(real_data, simulations):
    '''
    Compares real draft results to Monte Carlo simulations to evaluate fairness.

    Calculates fairness scores (z-scores), a p-value based on overall deviation, 
    and returns key data for further analysis and visualization.

    Parameters:
        real_data (pd.DataFrame): Actual draft results with 'Team' and 'Pick'.
        simulations (list of pd.DataFrame): Simulated draft results.

    Returns:
        dict: Contains fairness scores, p-value, real and simulated pick counts, 
              all simulations, and total deviation.
    '''
    # Count real outcomes
    real_counts = real_data.groupby(['Team', 'Pick']).size().unstack(fill_value=0)
    
    # Process simulations - ensure consistent indexing
    sim_dfs = []
    for sim in simulations:
        counts = sim.groupby(['Team', 'Pick']).size().unstack(fill_value=0)
        counts = counts.reindex(columns=real_counts.columns, fill_value=0)
        counts = counts.reindex(index=real_counts.index, fill_value=0)
        sim_dfs.append(counts)
    
    # Calculate mean simulated counts
    sim_counts = pd.concat(sim_dfs).groupby(level=0).mean()
    
    # Calculate fairness scores (lower = more suspicious)
    with np.errstate(divide='ignore', invalid='ignore'):
        fairness_scores = (real_counts - sim_counts) / np.sqrt(sim_counts)
        fairness_scores = fairness_scores.fillna(0)  # Handle 0/0 cases
    
    # Calculate p-value - convert to numpy arrays for comparison
    real_arr = real_counts.values
    sim_mean_arr = sim_counts.values
    
    total_deviation = np.nansum((real_arr - sim_mean_arr)**2 / (sim_mean_arr + 1e-10))

    # count cases where simulated results were less probable than actual results
    extreme_counts = 0
    for sim in sim_dfs:
        sim_arr = sim.values
        deviation = np.nansum((sim_arr - sim_mean_arr)**2 / (sim_mean_arr + 1e-10))
        if deviation >= total_deviation:
            extreme_counts += 1
    
    p_value = extreme_counts / len(sim_dfs)
    
    return {
    'fairness_scores': fairness_scores,
    'p_value': p_value,
    'real_counts': real_counts,
    'sim_counts': sim_counts,
    'sim_dfs': sim_dfs,
    'total_deviation': total_deviation
    }
````

## How the p-value is Calculated
The p-value is calculated by counting how many simulations resulted in a more extreme (less likely) outcome than what actually occurred, then dividing that count by the total number of simulations.

````python
# execute simulations
simulations_since_1990 = run_multi_year_simulation(yearly_odds)

# feed simulation results and real results into the comparison function
results = analyze_results(real_data, simulations_since_1990)
fairness_scores = results['fairness_scores']
p_value = results['p_value']
real_counts = results['real_counts']
sim_counts = results['sim_counts']
sim_dfs = results['sim_dfs']
total_deviation = results['total_deviation']
````

````python
print(f"Overall fairness p-value: {p_value:.4f}")
print("\nMost suspicious results:")
print(fairness_scores.abs().stack().nlargest(100))
````

## Where Does the Actual NBA Lottery Outcome in the Simulated Distribution?
The histogram below shows the distribution of deviation scores across all Monte Carlo simulations. These scores reflect how far each simulation’s results deviate from the expected average. The red dashed line represents the deviation of the actual NBA draft outcome. 

````python
# Extract deviation values from all simulations
deviations = [
    np.nansum((sim.values - sim_counts.values) ** 2 / (sim_counts.values + 1e-10))
    for sim in sim_dfs
]

plt.figure(figsize=(10, 6))
sns.histplot(deviations, bins=100, kde=True)
plt.axvline(total_deviation, color='red', linestyle='--', linewidth=3, label='Actual Draft Deviation')
plt.title("Monte Carlo Distribution of Simulated NBA Drafts", fontweight='bold', fontsize=16)
plt.xlabel("Simulated Deviation", fontweight='bold', fontsize=13)
plt.xlim(300,650)
plt.ylabel("Frequency", fontweight='bold', fontsize=13)
plt.legend(prop={'weight':'bold', 'size':12})
plt.tight_layout()
plt.show()
````

````python
print(f'Actual Deviation: {round(total_deviation,2)}')

# count occurrences with higher deviation
cnt_more_xtrm = np.sum(np.array(deviations) > total_deviation)
print(f'Extreme Occurrences: {cnt_more_xtrm}')

# divide that by total simulations to compute p-value
pval = cnt_more_xtrm/100000
print(f'p = {pval}')
````

Out of 100,000 simulations, 11.66% produced deviations as extreme as (or more extreme than) the actual draft outcomes (p = 0.1166). Using a significance threshold of α = 0.05, we cannot reject the null hypothesis that the results are random. While notable outliers exist (e.g. teams winning with <2% odds), they fall within the realm of statistical possibility.

# Inconclusive Evidence of Foul Play
Our analysis finds no statistically significant evidence to conclude that the NBA draft lottery is rigged. While the Mavericks’ 1.8%-probability win in 2025 raised eyebrows—especially following the controversial Dončić trade—the deviation from expected outcomes (*p* = 0.1166) falls within the range of random chance.

**Key Takeaways**
- Absence of Proof ≠ Proof of Absence:
- We fail to reject the null hypothesis, but this doesn’t prove the lottery was fair—only that its outcomes align with statistical expectations under randomness.
- Rigging could theoretically exist but be undetectable at this sample size or sophistication level.
- Rare events are not impossible - over a long period of time they will occur and draw skepticism

The NBA isn't just a sports league - it's an entertainment empire. And what's more entertaining than a team trading its superstar only to immediately land the next generational talent? While our numbers can't confirm foul play, they reveal why fans question its fairness: the league benefits most when reality plays out like compelling fiction. The Mavericks' improbable win feels meaningful because our brains are wired to connect dots, especially when they form dramatic narratives. The draft's drama may be precisely why its integrity will always be questioned.
